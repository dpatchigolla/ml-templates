```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_theme(style="darkgrid", color_codes=True)
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

### Bar Plot

```python
sns.set_style("whitegrid")
```

```python
ax = sns.barplot(x="day", y="total_bill", data=tips)
```

```python

def annotate_bar_plot(bar_plot, x_offset=0.07, y_offset=0.0002):
    for n, patch in enumerate(bar_plot.patches):
        height = patch.get_height()
        bp.text(patch.get_x() + x_offset, height + y_offset, "{:.2f}".format(100*height))
```

```python

plt.subplots(figsize=(15, 5))
sns.set_style("white")
bp=sns.barplot(x="day", y="total_bill", hue="sex", data=tips,ci=None, palette=sns.color_palette("Set2"))
annotate_bar_plot(bp, 0.1, 0.2)
plt.legend(loc="best")
plt.xticks(rotation=60)
plt.title("Bar plot with label values");
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("year", "month", "passengers")
flights = flights.div(flights.sum(axis=1), axis=0)
```

```python
flights.head()
```

```python
ax = sns.heatmap(flights,
                 linewidths=0.5,
                 robust=True, 
                 annot_kws={"size":14}, 
                 annot=True,
                 fmt=".2f",
                 cmap="YlGnBu")
ax.tick_params(labelsize=14)
ax.figure.set_size_inches((12, 8))
ax.set_ylabel("Year")
ax.set_xlabel("Month")
ax.set_title("Monthly flight passenger distribution across years");
```
